```python
cols_to_replace=['education','gender','marital_status','occupation']
user_features[cols_to_replace]
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
```

```python
user_features[cols_to_replace]=user_features[cols_to_replace].apply(lambda col:le.fit_transform(col))
user_features
```

```python
from sklearn.neighbors import KNeighborsClassifier, NearestNeighbors

model=KNeighborsClassifier(n_neighbors=10, metric="cosine")
model
```

```python
cols_to_replace=['age','income','education','gender','marital_status','occupation']
```

```python
x=user_features[cols_to_replace]
x
```

```python
y=user_features['user_id']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=0);
```
